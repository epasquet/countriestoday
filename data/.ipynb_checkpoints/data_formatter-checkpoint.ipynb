# ETAPE 1 : Selection de variables du factbook.csv

### Et conception d'indicateurs

```python
df = pd.read_csv("factbook_converti_float_ajout_col_complementaires.csv")
```

```python
output = pd.DataFrame()
output["Country"] = df["Country"]
```

### Electricity surplus

```python
selection = df[["Population", "Electricity - consumption(kWh)", "Electricity - production(kWh)"]]
selection["Electricity surplus"] = selection["Electricity - production(kWh)"] -\
                                   selection["Electricity - consumption(kWh)"]
selection["Electricity surplus (per capita)"] = selection["Electricity surplus"]\
                                                .divide(selection["Population"])
```

```python
output["Electricity surplus (per capita)"] = selection["Electricity surplus (per capita)"]
```

### kms d'autoroute par km2 et par habitant

```python
selection = df[["Population", "Country", "Area(sq km)", "Highways(km)"]]
selection["Highways(km) (per km2.capita)"] = selection["Highways(km)"]\
                                             .divide(selection["Area(sq km)"])\
                                             .divide(selection["Population"])
```

```python
output["Highways(km) (per km2.capita)"] = selection["Highways(km) (per km2.capita)"]
```

### Deficit pétrole et dépendance par rapport aux réserves de pétrole

```python
selection = df[["Population", "Oil - consumption(bbl/day)",\
                "Oil - production(bbl/day)", "Oil - proved reserves(bbl)"]]
selection["Oil deficit"] = selection["Oil - consumption(bbl/day)"] -\
                              selection["Oil - production(bbl/day)"]
selection["Oil deficit (per capita)"] = selection["Oil deficit"] / selection["Population"]
selection["Oil dependency (per capita)"] = selection["Population"] / selection["Oil - proved reserves(bbl)"]
```

```python
output["Oil deficit (per capita)"] = selection["Oil deficit (per capita)"]
output["Oil dependency (per capita)"] = selection["Oil dependency (per capita)"]
```

```python
selection = df[["Population", "Oil - consumption(bbl/day)",\
                "Oil - production(bbl/day)", "Oil - proved reserves(bbl)"]]
selection["Oil deficit"] = selection["Oil - consumption(bbl/day)"] -\
                              selection["Oil - production(bbl/day)"]
selection["Oil deficit (per capita)"] = selection["Oil deficit"] / selection["Population"]
selection["Oil dependency (per capita)"] = selection["Population"] / selection["Oil - proved reserves(bbl)"]
```

### PIB, indice de compétitivité à partir du PIB par habitant, de l'inflation, de l'investissement

### Binnage des dataframes CUSTOM

##### INFLATION : Une bonne inflation est entre 0 et 3%. Au-dessous de 0, on dégrade très vite le score. Au dessus de 0, on dégrade moins vite.

```python
inflation = output["Inflation rate(%)(consumer prices)"].copy()
```

```python
inflation[inflation < -3.] = -4.
inflation[(inflation >= -3.) & (inflation < -1.)] = -2.
inflation[(inflation >= -1.) & (inflation < 0.)] = -1.
inflation[(inflation >= 0.) & (inflation < 1.)] = 0.
inflation[(inflation >= 1.) & (inflation < 2.)] = 1.
inflation[(inflation >= 2.) & (inflation < 3.)] = 2.
inflation[(inflation >= 3.) & (inflation < 5.)] = 4.
inflation[(inflation >= 5.) & (inflation < 10.)] = 7.
inflation[inflation >= 10.] = 10.
```

```python
inflation[inflation== -4.] = 1.
inflation[inflation == -2.] = 2.
inflation[inflation == -1.] = 3.
inflation[inflation == 0.] = 4.
inflation[inflation == 1.] = 5.
inflation[inflation == 2.] = 4.
inflation[inflation == 4.] = 3.
inflation[inflation == 7.] = 2.
inflation[inflation == 10.] = 1.
```

```python
binned["Inflation rate(%)(consumer prices)"] = inflation
```

# FILLNA avec 0.5

```python
binned = binned.fillna(0.5)
```

# Ajout de la colonne SCORE - moyenne des bins

```python
binned["Score"] = binned.mean(axis=1)
```

```python
binned["Score"] -= binned["Score"].min()
binned["Score"] /= binned["Score"].max()
```
